```python
import os
import sys
import numpy as np
import seaborn as sns
%pylab inline
```

```python
from __future__ import print_function

# all the frame positions of every segment of the rav4 
# have been saved in this array.

f = open('frame_positions_rav4', 'rb')
frame_positions_rav4 = np.load(f, allow_pickle=True, encoding="bytes")

# 378 segments, with about 1200 ECEF positions in each segment
print(len(frame_positions_rav4), frame_positions_rav4[0].shape)
```

```python
sys.path.append('..')
sys.path.append('../utils')
from utils.benchmarks import get_altitude_errors
altitude_errors = get_altitude_errors(frame_positions_rav4)
```

```python
figsize(12,12);
sns.set_style("whitegrid")
sns.distplot(np.concatenate(altitude_errors), bins=arange(-10,10,.3), norm_hist=True, kde=False);
title('Altitude error distribution of rav4 positions', fontsize=25);
legend(fontsize=25);
xlabel('altitude error (m)', fontsize=18);
ylabel('density', fontsize=18);
```
